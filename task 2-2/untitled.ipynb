```
import pandas as pd
import matplotlib.pyplot as plt
from IPython import display
df = pd.read_csv('medical appointment dataset.csv')
df_before = df.copy()
bef_cle=df_before.shape
print("before cleaning dataset:",bef_cle)
missing_val=df.isnull().sum()
print(missing_val)
missing_val.plot(kind='bar',color='red')
plt.title("missing values graph")
plt.xlabel("columns")
plt.ylabel("missing values")
plt.show()
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(numerical_cols)
df[numerical_cols]=df[numerical_cols].fillna(df[numerical_cols].mean)
print(df[numerical_cols].isnull().sum())
categorical_cols=df.select_dtypes(include=['object']).columns
print(categorical_cols)
df[categorical_cols]=df[categorical_cols].fillna(df[categorical_cols].mode)
print(df[categorical_cols].isnull().sum())
df = df.loc[:, df.isnull().mean() < 0.7]

aft_cle = df.shape
print("\nAfter cleaning dataset shape:", aft_cle)

print("Validating dataset:")
print("\nMissing values after cleaning:")
print(df.isnull().sum())
print("Data types:")
print(df.dtypes)
print("Dataset info:")
df.info()
aft_cle = df.shape
print("\nAfter cleaning dataset shape:", aft_cle)
print("before vs after dataset cleaning")
print("Rows before cleaning:", df_before.shape[0])
print("Rows after cleaning :", df.shape[0])
print("Columns before cleaning:", df_before.shape[1])
print("Columns after cleaning :", df.shape[1])





```
